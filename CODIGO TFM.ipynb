Usa regresiÃ³n logÃ­stica con SMOTE.

Si el target es B1, entrena con aÃ±os previos.

Si el target es B2, entrena con aÃ±os previos + B1 del 2025.

Features segÃºn la estrategia:

M1 â†’ M1 y M3

M2 â†’ M1, M2 y M4

M3 â†’ M1 y M3

M4 â†’ M1, M2 y M4

Se puede extender fÃ¡cilmente a B3 y B4 en el futuro.

```python
# ============================================
# REGRESIÃ“N LOGÃSTICA - PREDICCIÃ“N M1â€“M4 (B1â€“B2)
# Estrategia: competencias especÃ­ficas + bimestres previos
# ============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    classification_report, f1_score, precision_score, recall_score,
    accuracy_score, confusion_matrix, ConfusionMatrixDisplay
)
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline

# =======================
# 0. Cargar datos
# =======================
df = pd.read_csv("/content/DATAMAT012025.csv")

# =======================
# 1. Definir estrategia de features por competencia
# =======================
estrategia = {
    "M1": ["M1", "M3"],
    "M2": ["M1", "M2", "M4"],
    "M3": ["M1", "M3"],
    "M4": ["M1", "M2", "M4"]
}
bimestres = ["B1", "B2", "B3", "B4"]

# =======================
# 2. Pipeline de regresiÃ³n logÃ­stica
# =======================
pipe = ImbPipeline([
    ('scaler', StandardScaler()),
    ('smote', SMOTE(random_state=42)),
    ('clf', LogisticRegression(solver='liblinear', max_iter=1000))
])

param_grid = {
    'clf__C': [0.01, 0.1, 1, 10],
    'clf__penalty': ['l1','l2']
}
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# =======================
# 3. Loop de predicciÃ³n
# =======================
resultados = []

for comp, feats in estrategia.items():
    for bim in ["B1","B2"]:  # solo B1 y B2
        target_2024 = f"{comp}_{bim}_2024"
        target_2025 = f"{comp}_{bim}_2025"

        if target_2024 not in df.columns or target_2025 not in df.columns:
            continue

        print("="*60)
        print(f"   {comp} - {bim}")
        print("="*60)

        # ---------------- VALIDACIÃ“N 2024 ----------------
        features = [f"{f}_{b}" for f in feats for b in bimestres]

        train_2024 = [f"{feat}_{year}" for feat in features for year in ["2021","2022","2023"]]
        X_train_2024 = df[train_2024]
        y_train_2024 = df[target_2024]

        grid_2024 = GridSearchCV(pipe, param_grid, scoring='f1', cv=cv, n_jobs=-1)
        grid_2024.fit(X_train_2024, y_train_2024)

        # Mejor umbral en 2024
        y_proba_2024 = grid_2024.predict_proba(X_train_2024)[:,1]
        thresholds = np.linspace(0,1,101)
        f1_scores = [f1_score(y_train_2024, (y_proba_2024 >= t).astype(int)) for t in thresholds]
        best_t = thresholds[np.argmax(f1_scores)]

        print("Mejores hiperparÃ¡metros:", grid_2024.best_params_)
        print("Umbral Ã³ptimo (mÃ¡x F1 en 2024):", round(best_t,2))
        print("\nMÃ©tricas 2024:")
        print(classification_report(y_train_2024, (y_proba_2024 >= best_t).astype(int)))

        # ---------------- PREDICCIÃ“N 2025 ----------------
        # Si es B1 â†’ usar solo aÃ±os anteriores
        # Si es B2 â†’ incluir B1 del 2025
        if bim == "B1":
            train_2025 = [f"{feat}_{year}" for feat in features for year in ["2022","2023","2024"]]
        elif bim == "B2":
            train_2025 = [f"{feat}_{year}" for feat in features for year in ["2022","2023","2024"]] \
                         + [f"{feat}_2025" for feat in features if "_B1" in feat]

        X_train_2025 = df[train_2025]
        y_train_2025 = df[target_2025]

        grid_2025 = GridSearchCV(pipe, param_grid, scoring='f1', cv=cv, n_jobs=-1)
        grid_2025.fit(X_train_2025, y_train_2025)

        y_proba_2025 = grid_2025.predict_proba(X_train_2025)[:,1]
        y_pred_2025 = (y_proba_2025 >= best_t).astype(int)

        # MÃ©tricas
        precision = precision_score(y_train_2025, y_pred_2025)
        recall = recall_score(y_train_2025, y_pred_2025)
        f1 = f1_score(y_train_2025, y_pred_2025)
        acc = accuracy_score(y_train_2025, y_pred_2025)

        print("\nMÃ©tricas 2025:")
        print(classification_report(y_train_2025, y_pred_2025))

        aciertos = np.sum(y_pred_2025 == y_train_2025)
        fallos = np.sum(y_pred_2025 != y_train_2025)
        print(f"Aciertos: {aciertos}, Fallos: {fallos}, Exactitud: {acc:.2%}")

        # Matriz de confusiÃ³n
        cm = confusion_matrix(y_train_2025, y_pred_2025)
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0,1])
        disp.plot(cmap=plt.cm.Blues, values_format="d")
        plt.title(f"{comp}-{bim} | Matriz ConfusiÃ³n 2025")
        plt.show()

        # Guardar resultados
        resultados.append({
            "Competencia": comp,
            "Bimestre": bim,
            "Mejores_Params": grid_2024.best_params_,
            "Umbral": round(best_t,2),
            "Accuracy_2025": round(acc,4),
            "Precision_2025": round(precision,4),
            "Recall_2025": round(recall,4),
            "F1_2025": round(f1,4),
            "Aciertos": aciertos,
            "Fallos": fallos
        })

# =======================
# 4. Resumen final
# =======================
df_resultados = pd.DataFrame(resultados)
print("\n===== RESUMEN DE RESULTADOS =====")
print(df_resultados)
```

```python
# ============================================
# SVM - PREDICCIÃ“N TEMPORAL
# Competencias M1â€“M4 | Bimestres B1â€“B2
# LÃ³gica: cada competencia usa sus propias notas y las previas
# ============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.svm import SVC
from sklearn.metrics import (
    classification_report, f1_score, precision_score, recall_score,
    accuracy_score, confusion_matrix, ConfusionMatrixDisplay
)
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline

# =======================
# 0. Cargar datos
# =======================
df = pd.read_csv("/content/DATAMAT012025.csv")

# =======================
# 1. Definir competencias y lÃ³gica de features
# =======================
bimestres = ["B1", "B2", "B3", "B4"]

# LÃ³gica de features por competencia
features_dict = {
    "M1": ["M1", "M3"],
    "M2": ["M1", "M2"],
    "M3": ["M1", "M2", "M3"],
    "M4": ["M1", "M2", "M3", "M4"]
}

# =======================
# 2. ConfiguraciÃ³n del pipeline SVM
# =======================
pipe = ImbPipeline([
    ('scaler', StandardScaler()),
    ('smote', SMOTE(random_state=42)),
    ('clf', SVC(probability=True))
])

param_grid = {
    'clf__kernel': ['linear', 'rbf', 'poly'],
    'clf__C': [0.1, 1, 10],
    'clf__gamma': ['scale', 'auto']
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Para guardar resultados
resultados = []

# =======================
# 3. Loop competencias y bimestres
# =======================
for comp, feats in features_dict.items():
    for bim in ["B1", "B2"]:
        target_2024 = f"{comp}_{bim}_2024"
        target_2025 = f"{comp}_{bim}_2025"

        if target_2024 not in df.columns or target_2025 not in df.columns:
            continue

        print("="*60)
        print(f"   {comp} - {bim}")
        print("="*60)

        # Construir lista de features con bimestres
        feats_all = [f"{c}_{b}" for c in feats for b in bimestres]

        # Entrenamiento validaciÃ³n 2024 (aÃ±os previos: 2021â€“2023)
        train_2024 = [f"{f}_{y}" for f in feats_all for y in ["2021", "2022", "2023"]]
        X_train_2024 = df[train_2024]
        y_train_2024 = df[target_2024]

        # Entrenamiento real 2025 (aÃ±os previos: 2022â€“2024)
        train_2025 = [f"{f}_{y}" for f in feats_all for y in ["2022", "2023", "2024"]]
        X_train_2025 = df[train_2025]
        y_train_2025 = df[target_2025]

        # =======================
        # 3.1. ValidaciÃ³n con 2024
        # =======================
        grid_2024 = GridSearchCV(pipe, param_grid, scoring='f1', cv=cv, n_jobs=-1)
        grid_2024.fit(X_train_2024, y_train_2024)

        y_proba_2024 = grid_2024.predict_proba(X_train_2024)[:, 1]
        thresholds = np.linspace(0, 1, 101)
        f1_scores = [f1_score(y_train_2024, (y_proba_2024 >= t).astype(int)) for t in thresholds]
        best_t = thresholds[np.argmax(f1_scores)]

        print("Mejores HiperparÃ¡metros 2024:", grid_2024.best_params_)
        print("Umbral Ã³ptimo 2024:", round(best_t, 2))

        # =======================
        # 3.2. EvaluaciÃ³n real con 2025
        # =======================
        grid_2025 = GridSearchCV(pipe, param_grid, scoring='f1', cv=cv, n_jobs=-1)
        grid_2025.fit(X_train_2025, y_train_2025)

        y_proba_2025 = grid_2025.predict_proba(X_train_2025)[:, 1]
        y_pred_2025 = (y_proba_2025 >= best_t).astype(int)

        # MÃ©tricas
        precision = precision_score(y_train_2025, y_pred_2025)
        recall = recall_score(y_train_2025, y_pred_2025)
        f1 = f1_score(y_train_2025, y_pred_2025)
        acc = accuracy_score(y_train_2025, y_pred_2025)

        print("\nReporte clasificaciÃ³n 2025:")
        print(classification_report(y_train_2025, y_pred_2025))
        print(f"Aciertos: {np.sum(y_pred_2025 == y_train_2025)}, Fallos: {np.sum(y_pred_2025 != y_train_2025)}, Exactitud: {acc:.2%}")

        # Matriz de confusiÃ³n
        cm = confusion_matrix(y_train_2025, y_pred_2025)
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0,1])
        disp.plot(cmap=plt.cm.Blues, values_format="d")
        plt.title(f"{comp}-{bim} | Matriz ConfusiÃ³n 2025 (SVM)")
        plt.show()

        # Guardar resultados
        resultados.append({
            "Competencia": comp,
            "Bimestre": bim,
            "Mejores_Params": grid_2024.best_params_,
            "Umbral": round(best_t, 2),
            "Accuracy_2025": round(acc, 4),
            "Precision_2025": round(precision, 4),
            "Recall_2025": round(recall, 4),
            "F1_2025": round(f1, 4),
            "Aciertos": int(np.sum(y_pred_2025 == y_train_2025)),
            "Fallos": int(np.sum(y_pred_2025 != y_train_2025))
        })

# =======================
# 4. Tabla resumen
# =======================
df_resultados = pd.DataFrame(resultados)
print("\n===== RESUMEN DE RESULTADOS =====")
print(df_resultados)
```

```python
# ============================================
# REGRESIÃ“N LOGÃSTICA - PREDICCIÃ“N TEMPORAL AUTOMATIZADA
# Para M1_B1_2025, M2_B1_2025, M3_B1_2025 y M4_B1_2025
# ============================================

import pandas as pd
import numpy as np
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, f1_score, confusion_matrix, ConfusionMatrixDisplay
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline
import matplotlib.pyplot as plt

# =======================
# 0. Cargar datos
# =======================
df = pd.read_csv("/content/DATAMAT012025.csv")

# =======================
# 1. ConfiguraciÃ³n del modelo
# =======================
pipe = ImbPipeline([
    ('scaler', StandardScaler()),
    ('smote', SMOTE(random_state=42)),
    ('clf', LogisticRegression(solver='liblinear', max_iter=1000, random_state=42))
])

param_grid = {
    'clf__C': [0.01, 0.1, 1, 10],
    'clf__penalty': ['l1','l2']
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

competencias = ["M1_B1", "M2_B1", "M3_B1", "M4_B1"]

# =======================
# 2. Bucle por competencia
# =======================
for comp in competencias:
    print(f"\n\n{'='*20} Competencia: {comp} {'='*20}")

    # ---- ValidaciÃ³n 2024 ----
    train_2024 = [col for col in df.columns if any(y in col for y in ["2021","2022","2023"])]
    X_train_2024 = df[train_2024]
    y_train_2024 = df[f"{comp}_2024"]

    grid_2024 = GridSearchCV(pipe, param_grid, scoring='f1', cv=cv, n_jobs=-1)
    grid_2024.fit(X_train_2024, y_train_2024)

    print("\nMejores HiperparÃ¡metros 2024:", grid_2024.best_params_)

    # Umbral Ã³ptimo
    y_proba_2024 = grid_2024.predict_proba(X_train_2024)[:,1]
    thresholds = np.linspace(0,1,101)
    f1_scores = [f1_score(y_train_2024, (y_proba_2024 >= t).astype(int)) for t in thresholds]
    best_t = thresholds[np.argmax(f1_scores)]
    print("Umbral Ã³ptimo (mÃ¡x F1 en 2024):", round(best_t,2))

    y_pred_2024 = (y_proba_2024 >= best_t).astype(int)
    print("\nMÃ©tricas en 2024:")
    print(classification_report(y_train_2024, y_pred_2024, zero_division=0))

    # Matriz de confusiÃ³n 2024
    cm_2024 = confusion_matrix(y_train_2024, y_pred_2024)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm_2024, display_labels=[0,1])
    disp.plot(cmap="Blues")
    plt.title(f"Matriz de ConfusiÃ³n {comp} - 2024")
    plt.show()

    # ---- PredicciÃ³n 2025 ----
    train_2025 = [col for col in df.columns if any(y in col for y in ["2022","2023","2024"])]
    X_train_2025 = df[train_2025]
    y_train_2025 = df[f"{comp}_2025"]

    grid_2025 = GridSearchCV(pipe, param_grid, scoring='f1', cv=cv, n_jobs=-1)
    grid_2025.fit(X_train_2025, y_train_2025)

    y_proba_2025 = grid_2025.predict_proba(X_train_2025)[:,1]
    y_pred_2025 = (y_proba_2025 >= best_t).astype(int)

    print("\n==== EvaluaciÃ³n real 2025 ====")
    print(classification_report(y_train_2025, y_pred_2025, zero_division=0))

    # Conteo aciertos/fallos
    aciertos = np.sum(y_pred_2025 == y_train_2025)
    fallos = np.sum(y_pred_2025 != y_train_2025)
    print(f"Aciertos: {aciertos}, Fallos: {fallos}, Exactitud: {aciertos/len(y_train_2025):.2%}")

    # Matriz de confusiÃ³n 2025
    cm_2025 = confusion_matrix(y_train_2025, y_pred_2025)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm_2025, display_labels=[0,1])
    disp.plot(cmap="Greens")
    plt.title(f"Matriz de ConfusiÃ³n {comp} - 2025")
    plt.show()
```

```python
# ============================================
# SVM - PREDICCIÃ“N TEMPORAL
# TODAS LAS COMPETENCIAS (M1â€“M4), BIMESTRES (B1â€“B2)
# ============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.svm import SVC
from sklearn.metrics import (
    classification_report, f1_score, precision_score, recall_score,
    accuracy_score, confusion_matrix, ConfusionMatrixDisplay
)
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline

# =======================
# 0. Cargar datos
# =======================
df = pd.read_csv("/content/DATAMAT012025.csv")

# =======================
# 1. Definir competencias y bimestres
# =======================
competencias = ["M1", "M2", "M3", "M4"]
bimestres = ["B1", "B2"]

# Pipeline
pipe = ImbPipeline([
    ('scaler', StandardScaler()),
    ('smote', SMOTE(random_state=42)),
    ('clf', SVC(probability=True, random_state=42))
])

# HiperparÃ¡metros para SVM
param_grid = {
    'clf__C': [0.1, 1, 10],
    'clf__kernel': ['linear', 'rbf'],
    'clf__gamma': ['scale', 'auto']
}
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Para guardar resultados
resultados = []

# =======================
# 2. Loop sobre todas las competencias y bimestres
# =======================
for comp in competencias:
    for bim in bimestres:
        target_2024 = f"{comp}_{bim}_2024"
        target_2025 = f"{comp}_{bim}_2025"

        # Evitar errores si no existe la columna
        if target_2024 not in df.columns or target_2025 not in df.columns:
            continue

        print("="*60)
        print(f"   {comp} - {bim}")
        print("="*60)

        # ---------- VALIDACIÃ“N 2024 ----------
        train_2024 = [col for col in df.columns if any(y in col for y in ["2021","2022","2023"])]
        X_train_2024 = df[train_2024]
        y_train_2024 = df[target_2024]

        grid_2024 = GridSearchCV(pipe, param_grid, scoring='f1', cv=cv, n_jobs=-1)
        grid_2024.fit(X_train_2024, y_train_2024)

        # Probabilidades y mejor umbral
        y_proba_2024 = grid_2024.predict_proba(X_train_2024)[:,1]
        thresholds = np.linspace(0,1,101)
        f1_scores = [f1_score(y_train_2024, (y_proba_2024 >= t).astype(int)) for t in thresholds]
        best_t = thresholds[np.argmax(f1_scores)]

        print("Mejores HiperparÃ¡metros 2024:", grid_2024.best_params_)
        print("Umbral Ã³ptimo 2024:", round(best_t,2))

        # ---------- PREDICCIÃ“N 2025 ----------
        train_2025 = [col for col in df.columns if any(y in col for y in ["2022","2023","2024"])]
        X_train_2025 = df[train_2025]
        y_train_2025 = df[target_2025]

        grid_2025 = GridSearchCV(pipe, param_grid, scoring='f1', cv=cv, n_jobs=-1)
        grid_2025.fit(X_train_2025, y_train_2025)

        y_proba_2025 = grid_2025.predict_proba(X_train_2025)[:,1]
        y_pred_2025 = (y_proba_2025 >= best_t).astype(int)

        # MÃ©tricas
        precision = precision_score(y_train_2025, y_pred_2025)
        recall = recall_score(y_train_2025, y_pred_2025)
        f1 = f1_score(y_train_2025, y_pred_2025)
        acc = accuracy_score(y_train_2025, y_pred_2025)

        # Conteo aciertos/fallos
        aciertos = np.sum(y_pred_2025 == y_train_2025)
        fallos = np.sum(y_pred_2025 != y_train_2025)

        print("\nReporte clasificaciÃ³n 2025:")
        print(classification_report(y_train_2025, y_pred_2025))
        print(f"Aciertos: {aciertos}, Fallos: {fallos}, Exactitud: {acc:.2%}")

        # Matriz de confusiÃ³n
        cm = confusion_matrix(y_train_2025, y_pred_2025)
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0,1])
        disp.plot(cmap=plt.cm.Oranges, values_format="d")
        plt.title(f"{comp}-{bim} | Matriz ConfusiÃ³n 2025")
        plt.show()

        # Guardar en resultados
        resultados.append({
            "Competencia": comp,
            "Bimestre": bim,
            "Mejores_Params": grid_2024.best_params_,
            "Umbral": round(best_t,2),
            "Accuracy_2025": round(acc,4),
            "Precision_2025": round(precision,4),
            "Recall_2025": round(recall,4),
            "F1_2025": round(f1,4),
            "Aciertos": aciertos,
            "Fallos": fallos
        })

# =======================
# 3. Tabla resumen
# =======================
df_resultados = pd.DataFrame(resultados)
print("\n===== RESUMEN DE RESULTADOS =====")
print(df_resultados)
```

```python
# ============================================
# LIGHTGBM - PREDICCIÃ“N TEMPORAL (M1â€“M4, B1â€“B2)
# ============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.metrics import (
    classification_report, f1_score, precision_score, recall_score,
    accuracy_score, confusion_matrix, ConfusionMatrixDisplay
)
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline
from lightgbm import LGBMClassifier

# =======================
# 0. Cargar datos
# =======================
df = pd.read_csv("/content/DATAMAT012025.csv")

# =======================
# 1. Definir lÃ³gica de features
# =======================
competencias_dict = {
    "M1": ["M1"],
    "M2": ["M1", "M2"],
    "M3": ["M1", "M2", "M3"],
    "M4": ["M1", "M2", "M3", "M4"]
}
bimestres = ["B1", "B2"]

# =======================
# 2. ConfiguraciÃ³n pipeline
# =======================
pipe = ImbPipeline([
    ('smote', SMOTE(random_state=42)),
    ('clf', LGBMClassifier(random_state=42))
])

param_grid = {
    'clf__n_estimators': [100, 200],
    'clf__max_depth': [-1, 5, 10],
    'clf__learning_rate': [0.05, 0.1, 0.2],
    'clf__num_leaves': [31, 63]
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# =======================
# 3. Loop competencias/bimestres
# =======================
resultados = []

for comp, feats in competencias_dict.items():
    for bim in bimestres:
        target_2024 = f"{comp}_{bim}_2024"
        target_2025 = f"{comp}_{bim}_2025"

        if target_2024 not in df.columns or target_2025 not in df.columns:
            continue

        print("="*60)
        print(f"   {comp} - {bim}")
        print("="*60)

        # ConstrucciÃ³n de features (todos los bimestres previos al actual)
        idx_bim = int(bim[1])  # B1=1, B2=2
        feats_bim = [f"{f}_B{i}" for f in feats for i in range(1, idx_bim+1)]

        # Features entrenamiento validaciÃ³n (2021â€“2023 â†’ 2024)
        X_train_2024 = df[[f"{fb}_{year}" for fb in feats_bim for year in ["2021","2022","2023"]]]
        y_train_2024 = df[target_2024]

        # GridSearchCV en validaciÃ³n
        grid_2024 = GridSearchCV(pipe, param_grid, scoring='f1', cv=cv, n_jobs=-1)
        grid_2024.fit(X_train_2024, y_train_2024)

        # Ajuste de umbral Ã³ptimo en 2024
        y_proba_2024 = grid_2024.predict_proba(X_train_2024)[:,1]
        thresholds = np.linspace(0,1,101)
        f1_scores = [f1_score(y_train_2024, (y_proba_2024 >= t).astype(int)) for t in thresholds]
        best_t = thresholds[np.argmax(f1_scores)]

        print("Mejores hiperparÃ¡metros 2024:", grid_2024.best_params_)
        print("Umbral Ã³ptimo 2024:", round(best_t,2))

        # Features entrenamiento predicciÃ³n (2022â€“2024 â†’ 2025)
        X_train_2025 = df[[f"{fb}_{year}" for fb in feats_bim for year in ["2022","2023","2024"]]]
        y_train_2025 = df[target_2025]

        grid_2025 = GridSearchCV(pipe, param_grid, scoring='f1', cv=cv, n_jobs=-1)
        grid_2025.fit(X_train_2025, y_train_2025)

        y_proba_2025 = grid_2025.predict_proba(X_train_2025)[:,1]
        y_pred_2025 = (y_proba_2025 >= best_t).astype(int)

        # MÃ©tricas
        precision = precision_score(y_train_2025, y_pred_2025)
        recall = recall_score(y_train_2025, y_pred_2025)
        f1 = f1_score(y_train_2025, y_pred_2025)
        acc = accuracy_score(y_train_2025, y_pred_2025)

        print("\nReporte clasificaciÃ³n 2025:")
        print(classification_report(y_train_2025, y_pred_2025))
        print(f"Aciertos: {(y_pred_2025 == y_train_2025).sum()}, Fallos: {(y_pred_2025 != y_train_2025).sum()}, Exactitud: {acc:.2%}")

        # Matriz de confusiÃ³n
        cm = confusion_matrix(y_train_2025, y_pred_2025)
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0,1])
        disp.plot(cmap=plt.cm.Blues, values_format="d")
        plt.title(f"{comp}-{bim} | Matriz ConfusiÃ³n 2025 (LightGBM)")
        plt.show()

        # Guardar resultados
        resultados.append({
            "Competencia": comp,
            "Bimestre": bim,
            "Mejores_Params": grid_2024.best_params_,
            "Umbral": round(best_t,2),
            "Accuracy_2025": round(acc,4),
            "Precision_2025": round(precision,4),
            "Recall_2025": round(recall,4),
            "F1_2025": round(f1,4)
        })

# =======================
# 4. Tabla resumen
# =======================
df_resultados = pd.DataFrame(resultados)
print("\n===== RESUMEN DE RESULTADOS =====")
print(df_resultados)
```
