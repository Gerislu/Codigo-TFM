REGRESIÓN LOGÍSTICA CON B1 Y B2

```python
# ============================================
# REGRESIÓN LOGÍSTICA - PREDICCIÓN M1–M4 (B1–B2)
# Estrategia: competencias específicas + bimestres previos
# ============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    classification_report, f1_score, precision_score, recall_score,
    accuracy_score, confusion_matrix, ConfusionMatrixDisplay
)
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline

# =======================
# 0. Cargar datos
# =======================
df = pd.read_csv("/content/DATAMAT012025.csv")

# =======================
# 1. Definir estrategia de features por competencia
# =======================
estrategia = {
    "M1": ["M1", "M3"],
    "M2": ["M1", "M2", "M4"],
    "M3": ["M1", "M3"],
    "M4": ["M1", "M2", "M4"]
}
bimestres = ["B1", "B2", "B3", "B4"]

# =======================
# 2. Pipeline de regresión logística
# =======================
pipe = ImbPipeline([
    ('scaler', StandardScaler()),
    ('smote', SMOTE(random_state=42)),
    ('clf', LogisticRegression(solver='liblinear', max_iter=1000))
])

param_grid = {
    'clf__C': [0.01, 0.1, 1, 10],
    'clf__penalty': ['l1','l2']
}
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# =======================
# 3. Loop de predicción
# =======================
resultados = []

for comp, feats in estrategia.items():
    for bim in ["B1","B2"]:  # solo B1 y B2
        target_2024 = f"{comp}_{bim}_2024"
        target_2025 = f"{comp}_{bim}_2025"

        if target_2024 not in df.columns or target_2025 not in df.columns:
            continue

        print("="*60)
        print(f"   {comp} - {bim}")
        print("="*60)

        # ---------------- VALIDACIÓN 2024 ----------------
        features = [f"{f}_{b}" for f in feats for b in bimestres]

        train_2024 = [f"{feat}_{year}" for feat in features for year in ["2021","2022","2023"]]
        X_train_2024 = df[train_2024]
        y_train_2024 = df[target_2024]

        grid_2024 = GridSearchCV(pipe, param_grid, scoring='f1', cv=cv, n_jobs=-1)
        grid_2024.fit(X_train_2024, y_train_2024)

        # Mejor umbral en 2024
        y_proba_2024 = grid_2024.predict_proba(X_train_2024)[:,1]
        thresholds = np.linspace(0,1,101)
        f1_scores = [f1_score(y_train_2024, (y_proba_2024 >= t).astype(int)) for t in thresholds]
        best_t = thresholds[np.argmax(f1_scores)]

        print("Mejores hiperparámetros:", grid_2024.best_params_)
        print("Umbral óptimo (máx F1 en 2024):", round(best_t,2))
        print("\nMétricas 2024:")
        print(classification_report(y_train_2024, (y_proba_2024 >= best_t).astype(int)))

        # ---------------- PREDICCIÓN 2025 ----------------
        if bim == "B1":
            train_2025 = [f"{feat}_{year}" for feat in features for year in ["2022","2023","2024"]]
        elif bim == "B2":
            train_2025 = [f"{feat}_{year}" for feat in features for year in ["2022","2023","2024"]] \
                         + [f"{feat}_2025" for feat in features if "_B1" in feat]

        X_train_2025 = df[train_2025]
        y_train_2025 = df[target_2025]

        grid_2025 = GridSearchCV(pipe, param_grid, scoring='f1', cv=cv, n_jobs=-1)
        grid_2025.fit(X_train_2025, y_train_2025)

        y_proba_2025 = grid_2025.predict_proba(X_train_2025)[:,1]
        y_pred_2025 = (y_proba_2025 >= best_t).astype(int)

        # Métricas
        precision = precision_score(y_train_2025, y_pred_2025)
        recall = recall_score(y_train_2025, y_pred_2025)
        f1 = f1_score(y_train_2025, y_pred_2025)
        acc = accuracy_score(y_train_2025, y_pred_2025)

        print("\nMétricas 2025:")
        print(classification_report(y_train_2025, y_pred_2025))

        aciertos = np.sum(y_pred_2025 == y_train_2025)
        fallos = np.sum(y_pred_2025 != y_train_2025)
        print(f"Aciertos: {aciertos}, Fallos: {fallos}, Exactitud: {acc:.2%}")

        # Matriz de confusión
        cm = confusion_matrix(y_train_2025, y_pred_2025)
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0,1])
        disp.plot(cmap=plt.cm.Blues, values_format="d")
        plt.title(f"{comp}-{bim} | Matriz Confusión 2025")
        plt.show()

        # Guardar resultados (incluyendo la matriz de confusión como texto)
        resultados.append({
            "Competencia": comp,
            "Bimestre": bim,
            "Mejores_Params": grid_2024.best_params_,
            "Umbral": round(best_t,2),
            "Accuracy_2025": round(acc,4),
            "Precision_2025": round(precision,4),
            "Recall_2025": round(recall,4),
            "F1_2025": round(f1,4),
            "Aciertos": aciertos,
            "Fallos": fallos,
            "Matriz_Confusion": cm.tolist()  # 👈 aquí se agrega la matriz en formato lista
        })

# =======================
# 4. Resumen final
# =======================
df_resultados = pd.DataFrame(resultados)
print("\n===== RESUMEN DE RESULTADOS =====")
print(df_resultados.to_string(index=False))
```

```python
# ============================================
# REGRESIÓN LOGÍSTICA - PREDICCIÓN M1–M4 (B3–B4) PARA 2025 + MATRICES DE CONFUSIÓN
# ============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    classification_report, f1_score, precision_score, recall_score,
    accuracy_score, confusion_matrix, ConfusionMatrixDisplay
)
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline

# =======================
# 0. Cargar datos
# =======================
df = pd.read_csv("/content/DATAMAT012025.csv")

# =======================
# 1. Estrategia de features
# =======================
estrategia = {
    "M1": ["M1", "M3"],
    "M2": ["M1", "M2", "M4"],
    "M3": ["M1", "M3"],
    "M4": ["M1", "M2", "M4"]
}
bimestres = ["B1", "B2", "B3", "B4"]  # para construir features históricos

# =======================
# 2. Pipeline de regresión logística
# =======================
pipe = ImbPipeline([
    ('scaler', StandardScaler()),
    ('smote', SMOTE(random_state=42)),
    ('clf', LogisticRegression(solver='liblinear', max_iter=1000))
])

param_grid = {
    'clf__C': [0.01, 0.1, 1, 10],
    'clf__penalty': ['l1', 'l2']
}
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# =======================
# 3. Loop de predicción B3 y B4
# =======================
resultados = []

for comp, feats in estrategia.items():
    for bim in ["B3", "B4"]:
        target_2024 = f"{comp}_{bim}_2024"

        if target_2024 not in df.columns:
            print(f"⚠️ No se encontró columna {target_2024}, se omite.")
            continue

        print("=" * 60)
        print(f"   {comp} - {bim}")
        print("=" * 60)

        # Construcción de features históricos (hasta el bimestre anterior)
        idx_bim = int(bim[1])
        features_historicas = [f"{f}_B{i}" for f in feats for i in range(1, idx_bim + 1)]

        # ---------------- ENTRENAMIENTO 2024 ----------------
        X_train_2024 = df[[f"{feat}_{year}" for feat in features_historicas for year in ["2021", "2022", "2023"]]]
        y_train_2024 = df[target_2024]

        grid_2024 = GridSearchCV(pipe, param_grid, scoring='f1', cv=cv, n_jobs=-1)
        grid_2024.fit(X_train_2024, y_train_2024)

        # Mejor umbral en 2024
        y_proba_2024 = grid_2024.predict_proba(X_train_2024)[:, 1]
        thresholds = np.linspace(0, 1, 101)
        f1_scores = [f1_score(y_train_2024, (y_proba_2024 >= t).astype(int)) for t in thresholds]
        best_t = thresholds[np.argmax(f1_scores)]

        print("Mejores hiperparámetros:", grid_2024.best_params_)
        print("Umbral óptimo (máx F1 en 2024):", round(best_t, 2))
        print("\nMétricas 2024:")
        y_pred_2024 = (y_proba_2024 >= best_t).astype(int)
        print(classification_report(y_train_2024, y_pred_2024))

        # === MATRIZ DE CONFUSIÓN ===
        cm = confusion_matrix(y_train_2024, y_pred_2024)
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0, 1])
        disp.plot(cmap='Blues', values_format='d')
        plt.title(f"Matriz de confusión {comp}-{bim}-2024")
        plt.show()

        print(f"Matriz de confusión {comp}-{bim}-2024:\n{cm}\n")

        # ---------------- PREDICCIÓN 2025 ----------------
        X_pred_2025 = df[[f"{feat}_{year}" for feat in features_historicas for year in ["2022", "2023", "2024"]]].copy()

        # Renombrar columnas temporalmente para coincidir con X_train_2024
        rename_dict = {}
        for feat in features_historicas:
            rename_dict[f"{feat}_2022"] = f"{feat}_2021"
            rename_dict[f"{feat}_2023"] = f"{feat}_2022"
            rename_dict[f"{feat}_2024"] = f"{feat}_2023"
        X_pred_2025.rename(columns=rename_dict, inplace=True)

        y_proba_2025 = grid_2024.predict_proba(X_pred_2025)[:, 1]
        y_pred_2025 = (y_proba_2025 >= best_t).astype(int)
        df[f"{comp}_{bim}_2025"] = y_pred_2025

        # Métricas sobre 2024 (entrenamiento)
        precision = precision_score(y_train_2024, y_pred_2024)
        recall = recall_score(y_train_2024, y_pred_2024)
        f1 = f1_score(y_train_2024, y_pred_2024)
        acc = accuracy_score(y_train_2024, y_pred_2024)

        resultados.append({
            "Competencia": comp,
            "Bimestre": bim,
            "Mejores_Params": grid_2024.best_params_,
            "Umbral": round(best_t, 2),
            "Accuracy_2024": round(acc, 4),
            "Precision_2024": round(precision, 4),
            "Recall_2024": round(recall, 4),
            "F1_2024": round(f1, 4),
            "MatrizConfusion": cm.tolist()
        })

        print(f"✅ Predicción {comp}-{bim}-2025 generada.\n")

# =======================
# 4. Resumen final
# =======================
df_resultados = pd.DataFrame(resultados)
print("\n===== RESUMEN DE RESULTADOS =====")
print(df_resultados)

# Guardar resultados y predicciones
df.to_csv("/content/DATAMAT012025_RegLog_Predicciones2025.csv", index=False)
df_resultados.to_csv("/content/Resultados_RegLog_B3B4.csv", index=False)
```

```python
# ============================================
# SVM - PREDICCIÓN TEMPORAL
# Competencias M1–M4 | Bimestres B1–B2
# Lógica: cada competencia usa sus propias notas y las previas
# ============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.svm import SVC
from sklearn.metrics import (
    classification_report, f1_score, precision_score, recall_score,
    accuracy_score, confusion_matrix, ConfusionMatrixDisplay
)
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline

# =======================
# 0. Cargar datos
# =======================
df = pd.read_csv("/content/DATAMAT012025.csv")

# =======================
# 1. Definir competencias y lógica de features
# =======================
bimestres = ["B1", "B2", "B3", "B4"]

# Lógica de features por competencia
features_dict = {
    "M1": ["M1", "M3"],
    "M2": ["M1", "M2"],
    "M3": ["M1", "M2", "M3"],
    "M4": ["M1", "M2", "M3", "M4"]
}

# =======================
# 2. Configuración del pipeline SVM
# =======================
pipe = ImbPipeline([
    ('scaler', StandardScaler()),
    ('smote', SMOTE(random_state=42)),
    ('clf', SVC(probability=True))
])

param_grid = {
    'clf__kernel': ['linear', 'rbf', 'poly'],
    'clf__C': [0.1, 1, 10],
    'clf__gamma': ['scale', 'auto']
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Para guardar resultados
resultados = []

# =======================
# 3. Loop competencias y bimestres
# =======================
for comp, feats in features_dict.items():
    for bim in ["B1", "B2"]:
        target_2024 = f"{comp}_{bim}_2024"
        target_2025 = f"{comp}_{bim}_2025"

        if target_2024 not in df.columns or target_2025 not in df.columns:
            continue

        print("="*60)
        print(f"   {comp} - {bim}")
        print("="*60)

        # Construir lista de features con bimestres
        feats_all = [f"{c}_{b}" for c in feats for b in bimestres]

        # Entrenamiento validación 2024 (años previos: 2021–2023)
        train_2024 = [f"{f}_{y}" for f in feats_all for y in ["2021", "2022", "2023"]]
        X_train_2024 = df[train_2024]
        y_train_2024 = df[target_2024]

        # Entrenamiento real 2025 (años previos: 2022–2024)
        train_2025 = [f"{f}_{y}" for f in feats_all for y in ["2022", "2023", "2024"]]
        X_train_2025 = df[train_2025]
        y_train_2025 = df[target_2025]

        # =======================
        # 3.1. Validación con 2024
        # =======================
        grid_2024 = GridSearchCV(pipe, param_grid, scoring='f1', cv=cv, n_jobs=-1)
        grid_2024.fit(X_train_2024, y_train_2024)

        y_proba_2024 = grid_2024.predict_proba(X_train_2024)[:, 1]
        thresholds = np.linspace(0, 1, 101)
        f1_scores = [f1_score(y_train_2024, (y_proba_2024 >= t).astype(int)) for t in thresholds]
        best_t = thresholds[np.argmax(f1_scores)]

        print("Mejores Hiperparámetros 2024:", grid_2024.best_params_)
        print("Umbral óptimo 2024:", round(best_t, 2))

        # =======================
        # 3.2. Evaluación real con 2025
        # =======================
        grid_2025 = GridSearchCV(pipe, param_grid, scoring='f1', cv=cv, n_jobs=-1)
        grid_2025.fit(X_train_2025, y_train_2025)

        y_proba_2025 = grid_2025.predict_proba(X_train_2025)[:, 1]
        y_pred_2025 = (y_proba_2025 >= best_t).astype(int)

        # Métricas
        precision = precision_score(y_train_2025, y_pred_2025)
        recall = recall_score(y_train_2025, y_pred_2025)
        f1 = f1_score(y_train_2025, y_pred_2025)
        acc = accuracy_score(y_train_2025, y_pred_2025)

        print("\nReporte clasificación 2025:")
        print(classification_report(y_train_2025, y_pred_2025))
        print(f"Aciertos: {np.sum(y_pred_2025 == y_train_2025)}, Fallos: {np.sum(y_pred_2025 != y_train_2025)}, Exactitud: {acc:.2%}")

        # Matriz de confusión
        cm = confusion_matrix(y_train_2025, y_pred_2025)
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0,1])
        disp.plot(cmap=plt.cm.Blues, values_format="d")
        plt.title(f"{comp}-{bim} | Matriz Confusión 2025 (SVM)")
        plt.show()

        # Guardar resultados (incluyendo matriz de confusión)
        resultados.append({
            "Competencia": comp,
            "Bimestre": bim,
            "Mejores_Params": grid_2024.best_params_,
            "Umbral": round(best_t, 2),
            "Accuracy_2025": round(acc, 4),
            "Precision_2025": round(precision, 4),
            "Recall_2025": round(recall, 4),
            "F1_2025": round(f1, 4),
            "Aciertos": int(np.sum(y_pred_2025 == y_train_2025)),
            "Fallos": int(np.sum(y_pred_2025 != y_train_2025)),
            "Matriz_Confusion": cm.tolist()  # 👈 se agrega la matriz de confusión como lista
        })

# =======================
# 4. Tabla resumen
# =======================
df_resultados = pd.DataFrame(resultados)
print("\n===== RESUMEN DE RESULTADOS =====")
print(df_resultados.to_string(index=False))
```

```python
# ============================================
# LIGHTGBM - PREDICCIÓN TEMPORAL (M1–M4, B3–B4) PARA 2025 + ROC/AUC + Importancia Features
# ============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.metrics import (
    f1_score, precision_score, recall_score, accuracy_score,
    roc_curve, auc
)
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline
from lightgbm import LGBMClassifier

# =======================
# 0. Cargar datos
# =======================
df = pd.read_csv("/content/DATAMAT012025.csv")

# =======================
# 1. Definir competencias y bimestres
# =======================
competencias_dict = {
    "M1": ["M1"],
    "M2": ["M1", "M2"],
    "M3": ["M1", "M2", "M3"],
    "M4": ["M1", "M2", "M3", "M4"]
}
bimestres = ["B3", "B4"]  # Solo B3 y B4

# =======================
# 2. Configuración pipeline
# =======================
pipe = ImbPipeline([
    ('smote', SMOTE(random_state=42)),
    ('clf', LGBMClassifier(random_state=42))
])

param_grid = {
    'clf__n_estimators': [100, 200],
    'clf__max_depth': [-1, 5, 10],
    'clf__learning_rate': [0.05, 0.1, 0.2],
    'clf__num_leaves': [31, 63]
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# =======================
# 3. Loop competencias/bimestres
# =======================
resultados = []

for comp, feats in competencias_dict.items():
    for bim in bimestres:
        target_2024 = f"{comp}_{bim}_2024"

        if target_2024 not in df.columns:
            print(f"⚠️  No se encontró columna {target_2024}, se omite.")
            continue

        print("="*60)
        print(f"   {comp} - {bim}")
        print("="*60)

        # Construcción de features
        idx_bim = int(bim[1])  # B3=3, B4=4
        feats_bim = [f"{f}_B{i}" for f in feats for i in range(1, idx_bim+1)]

        # =======================
        # ENTRENAMIENTO PARA 2024
        # =======================
        X_train_2024 = df[[f"{fb}_{year}" for fb in feats_bim for year in ["2021","2022","2023"]]]
        y_train_2024 = df[target_2024]

        grid_2024 = GridSearchCV(pipe, param_grid, scoring='f1', cv=cv, n_jobs=-1)
        grid_2024.fit(X_train_2024, y_train_2024)

        # Ajuste de umbral óptimo
        y_proba_2024 = grid_2024.predict_proba(X_train_2024)[:,1]
        thresholds = np.linspace(0,1,101)
        f1_scores = [f1_score(y_train_2024, (y_proba_2024 >= t).astype(int)) for t in thresholds]
        best_t = thresholds[np.argmax(f1_scores)]

        print("Mejores hiperparámetros:", grid_2024.best_params_)
        print("Umbral óptimo:", round(best_t,2))

        # =======================
        # Curva ROC / AUC 2024
        # =======================
        fpr, tpr, _ = roc_curve(y_train_2024, y_proba_2024)
        roc_auc = auc(fpr, tpr)

        plt.figure(figsize=(6,5))
        plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
        plt.plot([0,1], [0,1], color='red', lw=2, linestyle='--')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'ROC - {comp}-{bim}-2024')
        plt.legend(loc="lower right")
        plt.show()

        # =======================
        # Importancia de features LightGBM
        # =======================
        clf_best = grid_2024.best_estimator_.named_steps['clf']
        importances = clf_best.feature_importances_
        feat_names = X_train_2024.columns

        feat_imp_df = pd.DataFrame({
            'Feature': feat_names,
            'Importancia': importances
        }).sort_values(by='Importancia', ascending=False)

        plt.figure(figsize=(8,6))
        sns.barplot(x='Importancia', y='Feature', data=feat_imp_df.head(20), palette='viridis')
        plt.title(f'Importancia Features - {comp}-{bim}-2024')
        plt.tight_layout()
        plt.show()

        # =======================
        # Predicción para 2025
        # =======================
        X_pred_2025 = df[[f"{fb}_{year}" for fb in feats_bim for year in ["2022","2023","2024"]]]
        y_proba_2025 = grid_2024.predict_proba(X_pred_2025)[:,1]
        y_pred_2025 = (y_proba_2025 >= best_t).astype(int)
        df[f"{comp}_{bim}_2025"] = y_pred_2025

        # Métricas sobre 2024 (opcional)
        precision = precision_score(y_train_2024, (y_proba_2024 >= best_t).astype(int))
        recall = recall_score(y_train_2024, (y_proba_2024 >= best_t).astype(int))
        f1 = f1_score(y_train_2024, (y_proba_2024 >= best_t).astype(int))
        acc = accuracy_score(y_train_2024, (y_proba_2024 >= best_t).astype(int))

        resultados.append({
            "Competencia": comp,
            "Bimestre": bim,
            "Mejores_Params": grid_2024.best_params_,
            "Umbral": round(best_t,2),
            "Accuracy_2024": round(acc,4),
            "Precision_2024": round(precision,4),
            "Recall_2024": round(recall,4),
            "F1_2024": round(f1,4)
        })

        print(f"✅ Predicción {comp}-{bim}-2025 generada.\n")

# =======================
# 4. Resumen y exportación
# =======================
df_resultados = pd.DataFrame(resultados)
print("\n===== RESUMEN DE RESULTADOS =====")
print(df_resultados)

# Guardar resultados y predicciones
df.to_csv("/content/DATAMAT012025_Predicciones2025.csv", index=False)
df_resultados.to_csv("/content/Resultados_LightGBM_B3B4.csv", index=False)
```
